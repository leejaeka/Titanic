```python
cabins = df['Cabin'].unique()
```

```python
[str(i) for i in cabins]
cabins[0] = 'NAN'
cabins.sort()
```

```python
cabins
```

```python
df['Cabin'] = df['Cabin'].astype('category')
df['Cabin'].cat.set_categories(cabins, ordered=True, inplace=True)
```

```python
X_train = df.drop("Survived",axis=1)
X_train = df.drop("Name",axis=1)
y_train = df["Survived"]
```

## Let's do a quick random forest


```python
param_grid = [
    {'n_estimators':[10,50,100,500,1000], 'max_leaf_nodes':[16,32,64], 'n_jobs':[-1]}
]
```

```python
from sklearn.model_selection import GridSearchCV
rf = RandomForestRegressor()
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='neg_mean_squared_error')
```

```python
grid_search.fit(X_train, y_train)
```
